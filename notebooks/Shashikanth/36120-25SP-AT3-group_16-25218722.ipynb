## D. Feature Selection


### D.1 Approach "Correlation Analysis"


```python
# List of numeric features
num_features =["open", "high", "low", "close", "volume", "marketCap"]
```

```python
# Compute daily percent change for features
df_pct = df[num_features].pct_change()

# Compute next-day percent change for the target
df_pct['high_next_day_per'] = df['high'].pct_change().shift(-1)

# Drop rows with NaN values (first row from pct_change, last row from shift)
df_pct_clean = df_pct.dropna()
```

```python
# correlation of features
corr = df_pct_clean.corr()
# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title("Correlation Heatmap of Features", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# set the threshold for high correlation
threshold = 0.75
# Calculate the correlation matrix and filter for high correlations
corr_matrix = df_pct_clean.corr().abs()
high_corr = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
             .stack()
             .reset_index())
high_corr.columns = ['feature1', 'feature2', 'correlation']
high_corr = high_corr[high_corr['correlation'] > threshold]
print(high_corr)
```

```python
df=df.drop(columns=['marketCap'],axis=1)
```

## F. Feature Engineering

### F.1 New Features "Price-based features


```python
# Price differences
df['high_low_diff'] = df['high'] - df['low']
df['close_open_diff'] = df['close'] - df['open']
```

```python
# Rolling statistics
df['rolling_mean_3'] = df['close'].rolling(3).mean()
df['rolling_std_3'] = df['close'].rolling(3).std()
df['rolling_max_7'] = df['high'].shift(1).rolling(7).max()
df['rolling_min_7'] = df['low'].shift(1).rolling(7).min()
```

```python
feature_engineering_1_explanations = """
Price-based features capture short-term trends, volatility, and market momentum. 

- `high_low_diff` measures daily volatility.  
- `close_open_diff` reflects intraday momentum.  
- `rolling_mean_3` and `rolling_std_3` show short-term trends and volatility.  
- `rolling_max_7` and `rolling_min_7` capture weekly highs and lows.  

These features provide temporal and contextual insights that improve next-day high price predictions.
"""
```

```python
# Do not modify this code
print_tile(size="h3", key='feature_engineering_1_explanations', value=feature_engineering_1_explanations)
```

### G.1 Split Datasets

```python
# --- Define split proportions ---
n = len(df)
train_size = int(n * 0.7)   # 70% for training
val_size = int(n * 0.15)    # 15% for validation
test_size = n - train_size - val_size  # remaining 15% for test

# --- Generate indices ---
train_index = np.arange(0, train_size)
val_index = np.arange(train_size, train_size + val_size)
test_index = np.arange(train_size + val_size, n)

X = df[final_features].values
y = df[target_name].values

# --- Split the data ---
X_train, y_train = X[train_index], y[train_index]
X_val, y_val = X[val_index], y[val_index]
X_test, y_test = X[test_index], y[test_index]

print("Train size:", len(X_train))
print("Validation size:", len(X_val))
print("Test size:", len(X_test))
```

```python
data_splitting_explanations = """
For time series data like cryptocurrency prices, preserving the chronological order is crucial to avoid data leakage and ensure realistic evaluation. 

We used a sequential split, where earlier data is used for training, followed by validation and then testing. This approach mirrors real-world forecasting, where models learn from past data to predict future outcomes. 

Random shuffling is avoided since it would mix future information into the training set, leading to overly optimistic results. Thus, a time-based split is the best strategy for this dataset.
"""
```

### J.2 Set Hyperparameters


```python
# Hyperparameter grid for RandomForestRegressor
param_grid = {
    'n_estimators': [100, 200, 300],       # number of trees
    'max_depth': [5, 10, 20, None],        # maximum depth of trees
    'min_samples_split': [2, 5, 10],       # minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],         # minimum samples at leaf node
    'max_features': ['auto', 'sqrt']       # number of features to consider at each split
}
```

```python
hyperparameters_selection_explanations = """
Tuning these hyperparameters is important to optimize the Random Forest Regressor's performance and prevent overfitting or underfitting.

- `n_estimators`: Increasing the number of trees generally improves performance and reduces variance, but with higher computational cost.  
- `max_depth`: Controls the maximum depth of each tree, balancing model complexity and generalization.  
- `min_samples_split` and `min_samples_leaf`: Prevent trees from creating nodes with too few samples, reducing overfitting.  
- `max_features`: Determines how many features to consider at each split, affecting diversity among trees and model bias-variance tradeoff.

By tuning these parameters, we aim to find the combination that provides the best predictive accuracy for the next-day high price while maintaining generalization to unseen data.
"""
```

### J.3 Fit Model

```python
# Import GridSearchCV from sklearn
from sklearn.model_selection import GridSearchCV,TimeSeriesSplit
```

```python
tscv = TimeSeriesSplit(n_splits=5)  # 5 folds

# GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error',
    cv=tscv,
    n_jobs=-1,
    verbose=1
)

# Fit the model
grid_search.fit(X_train, y_train)

# Best model and parameters
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
print("Best CV RMSE:", -grid_search.best_score_)  # convert from negative RMSE
```
